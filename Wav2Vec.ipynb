```python
import soundfile as sf
import torch
from datasets import load_dataset
from transformers import Wav2Vec2ForCTC, Wav2Vec2Processor
```

```python
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
train_data = pd.read_csv("/content/drive/MyDrive/Speech_text/train (1).csv")
```

```python
all_transcription = []
for label in labels:
    print(label)
    waves = [f for f in os.listdir(train_audio_path + '/'+ label) if f.endswith('.wav')]
    for wav in waves:
        waveform , sr = librosa.load(train_audio_path + '/' + label + '/' + wav, sr = 16000)
        sample_rate = sr
        input_values = processor(waveform, sampling_rate=sample_rate, return_tensors="pt").input_values  # Batch size 1
        logits = model(input_values).logits
        predicted_ids = torch.argmax(logits, dim=-1)
        transcription = processor.decode(predicted_ids[0])
        all_transcription.append(transcription)
```
